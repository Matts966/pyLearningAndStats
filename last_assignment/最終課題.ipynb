今回は、OpenAIGymのTaxi-v2を強化学習で実装するため、学習手法に以下の二つを用いた。

## SARSA
> ${\displaystyle Q(s_{t},a_{t})\leftarrow Q(s_{t},a_{t})+\alpha [r_{t}+\gamma Q(s_{t+1},a_{t+1})-Q(s_{t},a_{t})]}$

## Q
> $Q(s_{t},a)\leftarrow Q(s_{t},a)+\alpha \left[r_{{t+1}}+\gamma \max _{p}Q(s_{{t+1}},p)-Q(s_{t},a)\right]$

Sarsaに関しては、最後まで「全く動かない」という局所解（-200）に収束してしまい、うまくいかなかった。アルゴリズムのみをメソッドに切り出して何度かデバッグしたが、なかなかうまくいかなかったので、今回はQ学習のみで諦めた。ハイパーパラメータの問題か、アルゴリズム実装のバグだと考えられるが、理論的には最適解に収束するはずと聞いていた（理解が覚束ない）ので、気がかりだった。タクシーの移動が、毎回フィールド全体を描画しているため読みづらいので、カーソルを制御文字で戻して描画を行なった点は、見易く、気に入っている。全体を通して、破壊的な関数が多くなってしまい、テストが難しかったが、パラメータの変化に注目していくつかユニットテストを実行したところ、バグの解決に役立った。また、細かい点では、pickleはlambda関数を保存できないため、dillをpickleとして使う、というのは、かなり分かりづらかった。

```python
import taxi_driver
import driver_constants
import gym, os, sys
import gym.spaces
from time import sleep
import random
from matplotlib import pyplot as plt
env = gym.make('Taxi-v2')
constants = driver_constants.driver_constants()
t = taxi_driver.taxi_driver(env, try_count=15000, show_process=False, update_function=constants.Q)
t.learn()
t.show_learning_curve()
t = taxi_driver.taxi_driver(env, try_count=15000, show_process=False, update_function=constants.Sarsa)
t.learn()
t.show_learning_curve()
```
